```python
#Create Data Dictionary
class create_data_dictionary:

    def __init__(self):
        '''This class provides functions to quickly develop a data dictionary for your data set'''
        return None

    def make_my_data_dictionary(self, dataFrame):
        '''Create an initial data dictionary excluding definitions for meaning of features'''

        col_ = dataFrame.columns
        df_DataDict = {}

        for col in col_:
                df_DataDict[col] = {
                               'Type': str(df.dtypes[col]),
                               'Length': len(df[col]),
                               'Null_Count': sum(df[col].isna()),
                               'Size(Memory)': df.memory_usage()[col],
                               'Definition': str('')
                                }

        df_DD = pd.DataFrame(df_DataDict)

        return df_DD

    def define_data_meaning(self, df_data_dictionary):
        '''Quickly provide input regarding each columns meaning and transpose into a usable dictionary'''

        col_ = df_data_dictionary.columns
        d = 'Definition'

        for col in col_:
            df_data_dictionary[col][d] = input('Provide a data definition for {}'.format(col))

        df_data_dictionary = df_data_dictionary.transpose()

        return df_data_dictionary

    def update_dd_definition(self, df_data_dictionary, attribute):
        try:
            df_dd = df_data_dictionary.transpose()
            df_dd[attribute]['Definition'] = input('Provide a data definition for {}'.format(attribute))
            df_dd = df_dd.transpose()
            return df_dd
        except:
            print('Sorry, there was an error.  Check attribute name and try again')
```

```python
# Question 1
# Histograms of Resilience scores for athletes and non-athletes
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis, norm

# Define values to exclude
excluded_values = [999]  # Replace with actual values

# Filter out the excluded values and create a copy
filtered_df = df[~df['RES_TOTAL'].isin(excluded_values)].copy()

# Correct mapping of class codes
class_labels = {1: "Athlete", 2: "Non-Athlete"}

# Replace numeric codes with meaningful labels
filtered_df['Athlete/Non-Athlete'] = filtered_df['Athlete/Non-Athlete'].replace(class_labels)

# Define a good color palette
custom_palette = ["#4c72b0", "#dd8452"]  # Blue for athletes, orange for non-athletes

# Function to calculate and return summary statistics as a string
def get_summary_stats(data):
    mean = np.mean(data)
    std = np.std(data)
    skewness = skew(data)
    kurt = kurtosis(data)
    summary = (f"Mean: {mean:.2f}\n"
               f"SD: {std:.2f}\n"
               f"Skew: {skewness:.2f}\n"
               f"Kurt: {kurt:.2f}")
    return summary

# Create side-by-side histograms using subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Athlete histogram
athlete_data = filtered_df[filtered_df['Athlete/Non-Athlete'] == "Athlete"]['RES_TOTAL']
sns.histplot(athlete_data, color=custom_palette[0], kde=False, ax=axes[0], alpha=0.7, stat="density")

# Overlay normal curve
mean_athlete = np.mean(athlete_data)
std_athlete = np.std(athlete_data)
xmin, xmax = axes[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean_athlete, std_athlete)
axes[0].plot(x, y, 'k', linewidth=2, label='Normal Curve')

# Summary statistics
axes[0].set_title("Athletes")
axes[0].set_xlabel("Resilience Score")
axes[0].set_ylabel("Density")
axes[0].text(0.05, 0.95, get_summary_stats(athlete_data), 
             transform=axes[0].transAxes, fontsize=10, va='top', ha='left',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="white", edgecolor="black"))

# Legend farther down below the summary statistics
axes[0].legend(loc='upper left', bbox_to_anchor=(0.05, 0.75))

# Non-Athlete histogram
non_athlete_data = filtered_df[filtered_df['Athlete/Non-Athlete'] == "Non-Athlete"]['RES_TOTAL']
sns.histplot(non_athlete_data, color=custom_palette[1], kde=False, ax=axes[1], alpha=0.7, stat="density")

# Overlay normal curve
mean_non_athlete = np.mean(non_athlete_data)
std_non_athlete = np.std(non_athlete_data)
xmin, xmax = axes[1].get_xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean_non_athlete, std_non_athlete)
axes[1].plot(x, y, 'k', linewidth=2, label='Normal Curve')

# Summary statistics
axes[1].set_title("Non-Athletes")
axes[1].set_xlabel("Resilience Score")
axes[1].text(0.05, 0.95, get_summary_stats(non_athlete_data), 
             transform=axes[1].transAxes, fontsize=10, va='top', ha='left',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="white", edgecolor="black"))

# Legend farther down below the summary statistics
axes[1].legend(loc='upper left', bbox_to_anchor=(0.05, 0.75))

# Overall title
fig.suptitle("Histograms of Resilience Scores for Athletes and Non-Athletes with Normal Curves", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
#Question 2
# Histograms and descriptive statistics for resilience and mental health scores
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis, norm

# Define values to exclude
excluded_values = [999]  # Values to filter out

# Count how many records are filtered out due to missing values
initial_count = len(df)
filtered_df = df[(~df['RES_TOTAL'].isin(excluded_values)) & 
                 (~df['MHC-SF OVERALL'].isin(excluded_values))].copy()
filtered_count = initial_count - len(filtered_df)  # Number of filtered records

# Define a color palette
custom_palette = ["#4c72b0", "#dd8452"]  # Blue and orange

# Function to calculate and return summary statistics as a string
def get_summary_stats(data):
    count = len(data)
    mean = np.mean(data)
    std = np.std(data)
    skewness = skew(data)
    kurt = kurtosis(data)
    summary = (f"Count: {count}\n"
               f"Mean: {mean:.2f}\n"
               f"SD: {std:.2f}\n"
               f"Skew: {skewness:.2f}\n"
               f"Kurt: {kurt:.2f}")
    return summary

# Create side-by-side histograms using subplots with separate axes
fig, axes = plt.subplots(1, 2, figsize=(12, 7), sharey=False)  # Set sharey=False for separate vertical axes

# ====== RES_TOTAL (Resilience Scores) ======
res_data = filtered_df['RES_TOTAL']
sns.histplot(res_data, color=custom_palette[0], kde=False, ax=axes[0], alpha=0.7, stat="count")  # Use frequency

# Overlay normal curve
mean_res = np.mean(res_data)
std_res = np.std(res_data)
xmin, xmax = axes[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean_res, std_res) * len(res_data) * (xmax - xmin) / 20  # Adjust height for frequency
axes[0].plot(x, y, 'k', linewidth=2, label='Normal Curve')

# Summary statistics
axes[0].set_title("Resilience Scores")
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Frequency")  # Updated Y-axis label to Frequency
axes[0].text(0.05, 0.95, get_summary_stats(res_data), 
             transform=axes[0].transAxes, fontsize=10, va='top', ha='left',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="white", edgecolor="black"))
axes[0].legend(loc='upper left', bbox_to_anchor=(0.05, 0.75))


# ====== MHC-SF OVERALL (Mental Health Continuum Score) ======
mhc_data = filtered_df['MHC-SF OVERALL']
sns.histplot(mhc_data, color=custom_palette[1], kde=False, ax=axes[1], alpha=0.7, stat="count")  # Use frequency

# Overlay normal curve
mean_mhc = np.mean(mhc_data)
std_mhc = np.std(mhc_data)
xmin, xmax = axes[1].get_xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean_mhc, std_mhc) * len(mhc_data) * (xmax - xmin) / 20  # Adjust height for frequency
axes[1].plot(x, y, 'k', linewidth=2, label='Normal Curve')

# Summary statistics
axes[1].set_title("Mental Health Continuum Score")
axes[1].set_xlabel("Score")
axes[1].set_ylabel("Frequency")  # Updated Y-axis label to Frequency
axes[1].text(0.05, 0.95, get_summary_stats(mhc_data), 
             transform=axes[1].transAxes, fontsize=10, va='top', ha='left',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="white", edgecolor="black"))
axes[1].legend(loc='upper left', bbox_to_anchor=(0.05, 0.75))

# Overall title
fig.suptitle("Histograms of Resilience Scores and Mental Health Continuum Scores with Normal Curves", fontsize=16)

# Footnote showing the number of filtered records
plt.figtext(0.5, 0.01, f"Note: {filtered_count} records were filtered out due to missing values on either variable.",
            ha="center", fontsize=10, bbox=dict(boxstyle="round,pad=0.5", facecolor="white", edgecolor="black"))

plt.tight_layout(rect=[0, 0.03, 1, 1])  # Adjust layout to make space for footnote
plt.show()
```
